```python
%matplotlib inline
import os
import sys
cwd = os.getcwd()
sys.path.append(os.path.join(cwd, '..'))

import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from aye.train import Trainer
from aye.blob_analysis import region_props
```

```python
k = 0.4
n_green = 7
n_red = 7
s = 0.4
```

# Improved detection with both red and green

```python
red1 = Trainer('red', k=k, n_images=n_red, test_size=s, use_ind=False)
red1.train()
```

```python
red1.validate()
```

```python
red2 = Trainer('red', k=k, n_images=n_red, test_size=s, use_ind=True)
red2.train()
```

```python
red2.validate()
```

```python
green1 = Trainer('green', k=k, n_images=n_green, test_size=s, use_ind=False)
green1.train()
```

```python
green1.validate()
```

```python
green2 = Trainer('green', k=0.5, n_images=7, test_size=0.4, use_ind=True)
green2.train()
```

```python
s_v, bw = green2.validate()
```

```python
import cv2
from aye.visualization import draw_bboxes

blobs, bw_cleaned = region_props(bw)


disp = cv2.cvtColor(bw_cleaned, cv2.COLOR_GRAY2RGB)
good_blobs = []

for blob in blobs:
    if blob['extent'] > 0.5:
        good_blobs.append(blob)

draw_bboxes(disp, good_blobs['bbox'], (255, 0, 0), 2)

fig = plt.figure(figsize=(16, 16))
ax = fig.add_subplot(121)
ax.imshow(s_v.im_raw)
ax = fig.add_subplot(122)
ax.imshow(disp)
```
